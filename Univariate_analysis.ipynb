#  Diving into Amazon UK Product Insights

```python
#LOAD THE DATA
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import skew, kurtosis

df = pd.read_csv("/Users/annapisarek/Downloads/amz_uk_price_prediction_dataset.csv")
df.head()
```

## Part 1: Product Categories

#### Business Question: What are the most popular product categories on Amazon UK, and how do they compare in terms of listing frequency?

```python
frequency_table = df['category'].value_counts()
print(frequency_table)
```

#### Which are the top 5 most listed product categories?

```python
frequency_table_top_5 = df['category'].value_counts().head(5)
print(frequency_table_top_5)
```

```python
frequency_table_top_10 = df['category'].value_counts().head(10)
```

```python
# Create a bar chart
plt.figure(figsize=(10, 6))
frequency_table_top_10.plot(kind='bar')

# Add labels and title
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.title('Frequency of Categories')

# Display the chart
plt.show()
```

```python
# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(frequency_table_top_10, labels=frequency_table_top_10.index, autopct='%1.1f%%', startangle=90)

# Add a title
plt.title('Distribution of Categories')

# Display the chart
plt.show()
```

## Part 2: Delving into Product Pricing

#### How are products priced on Amazon UK, and are there specific price points or ranges that are more common?

```python
# Calculate the mean
mean_price = df['price'].mean()

# Calculate the median
median_price = df['price'].median()

# Calculate the mode 
mode_price = df['price'].mode()

# Display the results
print(f"Mean Price: {mean_price}")
print(f"Median Price: {median_price}")
print(f"Mode Price: {mode_price}")
```

##### The average price is just below 90 but the median is a lot lower, which means that there are some very expensive outliers in the database. That's why the 19.09 median is more reliable.

```python


# Create a histogram for the price column
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=30, edgecolor='black')

# Add labels and title
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Product Prices')

# Display the histogram
plt.show()
```

```python
##### The outliers make it difficult to visualise. We can use discretization for that.
```

```python
# Discretizing 'Price' into quartiles
df['price_q'] = pd.qcut(df['price'], q=4, labels=['Low', 'Medium', 'High', 'Highest'])

```

```python
import matplotlib.pyplot as plt

# Count the frequency of each quartile
quartile_counts = df['price_q'].value_counts()

# Create a bar chart
plt.figure(figsize=(10, 6))
quartile_counts.plot(kind='bar', color=['lightblue', 'lightgreen', 'orange', 'red'])

# Add labels and title
plt.xlabel('Price Quartiles')
plt.ylabel('Frequency')
plt.title('Frequency of Product Prices by Quartile')

# Display the chart
plt.xticks(rotation=0)  # Rotate x labels for better readability
plt.show()
```

```python
# Create a box plot for the price column
plt.figure(figsize=(10, 6))
plt.boxplot(df['price'], vert=False)

# Add labels and title
plt.xlabel('Price')
plt.title('Box Plot of Product Prices')

# Display the box plot
plt.show()
```

## Part 3: Unpacking Product Ratings

#### Measures of Centrality

```python
#Clean the DF
df_reviews = df[df['stars'] != 0.0]

# Calculate the mean
mean = df_reviews['stars'].mean()

# Calculate the median
median = df_reviews['stars'].median()

# Calculate the mode 
mode = df_reviews['stars'].mode()

# Display the results
print(f"Mean Rating: {mean}")
print(f"Median Rating: {median}")
print(f"Mode Rating: {mode}")
```

##### Most customer don't rate products, but if they do they rate it quite positively, mostly about 4.5


#### Measures of Dispersion

```python
df_reviews['stars'].describe()
```

```python
df['stars'].describe()
```

##### The ratings are consistent when they are given.

#### Shape of the Distribution:



```python


# Only for reviewed products

# Calculate skewness
skewness = skew(df_reviews['stars'])

# Calculate kurtosis
kurt = kurtosis(df_reviews['stars'])

# Display the results
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")

# Including missing reviews

import pandas as pd
from scipy.stats import skew, kurtosis

# Assuming df is your DataFrame and you have already cleaned it by removing 0.0 values

# Calculate skewness
skewness = skew(df['stars'])

# Calculate kurtosis
kurt = kurtosis(df['stars'])

# Display the results
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")

```
